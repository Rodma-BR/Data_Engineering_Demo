```python
import pyspark.sql.functions as F
import pyspark.sql.types as T
```

```python
df_customers = spark.table("momentum.bronze.customers")
```

```python
df_customers.display()
```

```python
df_customers_cleaned = df_customers.withColumns({"Gender_Label": F.when(F.col("Gender")==True, "H").otherwise("M"),
                                                 "Age" : (F.year(F.current_date()) - F.year(F.col("Birth_Date"))),
                                                 "Phone_Number" : F.concat(F.substring("Phone_Number", 1, 5), F.regexp_replace(F.substring("Phone_Number", 6, F.length("Phone_Number")), r'(\d)', 'X'))
                                                 })
```
