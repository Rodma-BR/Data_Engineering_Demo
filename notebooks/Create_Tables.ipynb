```python
import pyspark.sql.functions as F
import pyspark.sql.types as T
import datetime as dt
```

```python
access_schema = T.StructType([
    T.StructField("Visit_Id", T.StringType(), False),
    T.StructField("Member_Id", T.StringType(), False),
    T.StructField("Plan_Id", T.IntegerType(), False),
    T.StructField("Branch_Id", T.StringType(), False),
    T.StructField("Amount", T.StringType(), False),
    T.StructField("Date", T.DateType(), False)
])

members_schema = T.StructType([
    T.StructField("Member_Id", T.StringType(), False),
    T.StructField("Name", T.StringType(), False),
    T.StructField("Last_Name", T.StringType(), False),
    T.StructField("Gender", T.BooleanType(), False),
    T.StructField("Birth_Date", T.DateType(), True),
    .StructField("Lon", T.FloatType(), True),
    .StructField("Lat", T.FloatType(), True),
    T.StructField("Zipcode", T.IntegerType(), True),
    T.StructField("Phone_Number", T.StringType(), True),
    T.StructField("Updated_At", T.TimestampType(), False)
])

suscription_plan_schema = T.StructType([
    T.StructField("Plan_Id", T.StringType(), False),
    T.StructField("Plan_Name", T.StringType(), False),
    T.StructField("Plan_Cost", T.StringType(), False),
    T.StructField("Plan_Duration", T.IntegerType(), False),
    T.StructField("Plan_Start_Time", T.TimestampType(), False),
    T.StructField("Plan_End_Time", T.TimestampType(), True),
    T.StructField("Plan_Description", T.StringType(), False),
    T.StructField("Updated_At", T.TimestampType(), False)
])

suscription_payment_schema = T.StructType([
    T.StructField("Payment_Id", T.StringType(), False),
    T.StructField("Plan_Id", T.StringType(), False),
    T.StructField("Member_Id", T.StringType(), False),
    T.StructField("Payment_Amount", T.FloatType(), False),
    T.StructField("Payment_Time", T.TimestampType(), False),
    T.StructField("Plan_Expiration_Time", T.TimestampType(), True),
    T.StructField("Payment_Status", T.StringType(), False),
    T.StructField("Updated_At", T.TimestampType(), False)
])
```
