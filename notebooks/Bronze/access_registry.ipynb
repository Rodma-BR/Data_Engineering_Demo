```python
import pyspark.sql.functions as F
import pyspark.sql.types as T
import datetime as dt
```

```python
access_schema = T.StructType([
    T.StructField("Visit_Id", T.IntegerType(), False),
    T.StructField("Customer_Id", T.IntegerType(), False),
    T.StructField("Plan_Id", T.IntegerType(), False),
    T.StructField("Branch_Id", T.IntegerType(), False),
    T.StructField("Payment_Id", T.StringType(), False),
    T.StructField("Date", T.DateType(), False),
    T.StructField("Visit_Time", T.TimestampType(), False),
    T.StructField("Updated_At", T.TimestampType(), False)
])

# access_polars_schema = {"Visit_Id" : pl.Int64,
#                     "Customer_Id" : pl.Int64,
#                     "Plan_Id" : pl.Int64,
#                     "Branch_Id" : pl.Int8,
#                     "Payment_Id" : pl.String,
#                     "Date" : pl.Date,
#                     "Visit_Time" : pl.Datetime,
#                     "Updated_At" : pl.Datetime
# }
```

```python
df_access = df = spark.read.csv("/Volumes/momentum/bronze/external_files/access.csv", 
    header=True, 
    schema = access_schema
)
```
