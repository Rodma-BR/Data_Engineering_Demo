```python
import pyspark.sql.functions as F
import pyspark.sql.types as T
import datetime as dt
```

```python
access_plans_schema = T.StructType([
    T.StructField("Plan_Id", T.StringType(), False),
    T.StructField("Plan_Name", T.StringType(), False),
    T.StructField("Plan_Cost", T.StringType(), False),
    T.StructField("Plan_Duration", T.IntegerType(), False),
    T.StructField("Plan_Start_Time", T.TimestampType(), False),
    T.StructField("Plan_End_Time", T.TimestampType(), True),
    T.StructField("Plan_Description", T.StringType(), False),
    T.StructField("Updated_At", T.TimestampType(), False)
])
```

```python
suscriptions = [("D1", "Daily_Pass", 150, 1, dt.datetime(2026,1,1), dt.datetime(2026,3,23), "Valid for 24 hours", dt.datetime.now()),
                ("MT1", "Momentum_Pass", 650, 30, dt.datetime(2026,1,1), None, "Valid for 30 days and includes 5 daily passes", dt.datetime.now()),
                ("M1", "Monthly_Pass", 1200, 30, dt.datetime(2026,1,1), None,  "Valid for 30 days and includes unlimited access", dt.datetime.now()),
                ("Y1", "Yearly_Pass", 10000, 365, dt.datetime(2026,1,1), None,  "Valid for 365 days and includes unlimited access", dt.datetime.now()),
                ("D2", "Daily_Pass", 180, 1, dt.datetime(2026,3,23), None, "Valid for 24 hours", dt.datetime.now())
                ]

df_sc = spark.createDataFrame(suscriptions, schema = access_plans_schema)
df_sc.display()
```
