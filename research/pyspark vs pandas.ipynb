## Pandas vs Pyspark

Common functions for each module

```python
import pandas as pd
import pyspark.sql.functions as F
```

```python
path = "/Volumes/momentum/bronze/external_files/customers.csv"
df_sp = spark.read.csv(path, header = True)
df_pd = pd.read_csv(path)
#pandas_df = spark_df.toPandas()
```

```python
df_sp.display()
df_pd
```

```python
df_sp.select(["Name", "Gender"]).display()
df_pd[["Name", "Gender"]]
```

```python
df_sp.limit(3).display()
df_pd.head(3)
```

```python
df_sp.filter(F.col("Gender")==False).display()
df_pd.loc[df_pd["Gender"]==False]
```

```python
print(df_sp.count())
print(df_pd.shape[0])
```

```python
df_sp.groupBy("Zipcode").agg(F.count("*").alias("Total")).display()
df_pd.groupby("Zipcode").agg({"Zipcode": "count"})
```

```python
df_sp.withColumn("Gender_Label", F.when(F.col("Gender") == True, "H").otherwise("M")).display()
df_pd["Gender_Label"] = "H"
df_pd.loc[df_pd["Gender"] == False, "Gender_Label"] = "M"
df_pd
```
